## 3. ***Data Wrangling***

### Data Wrangling Code

**Handling Missing Values in the dataset**

````python
#calculating the mode of country column
hotel_df["country"].mode()
````

````python
# Fill missing values in 'children' with 0 (assuming missing means no children)
hotel_df['children'] = hotel_df['children'].fillna(0)

# Fill missing 'country' with the most common country(mode)
hotel_df['country'] = hotel_df['country'].fillna(hotel_df['country'].mode()[0])

# Fill missing 'agent' with 0 (assuming missing means no agent)
hotel_df[['agent']] = hotel_df[['agent']].fillna(0)

# droping the company coloumn as it  has highest null values
hotel_df.drop(['company'], axis=1, inplace=True)
````

**Converting Data Types for Optimization**

````python
# Convert categorical columns to 'category' type
categorical_cols = ['hotel', 'arrival_date_month', 'meal', 'market_segment',
                    'distribution_channel', 'reserved_room_type', 'assigned_room_type',
                    'deposit_type', 'customer_type', 'reservation_status']
hotel_df[categorical_cols] = hotel_df[categorical_cols].astype('category')

# Convert dates to datetime format
hotel_df['reservation_status_date'] = pd.to_datetime(hotel_df['reservation_status_date'])

# Convert agent & children columns to integer (since we replaced NaN with 0)
hotel_df[['agent', 'children']] = hotel_df[['agent', 'children']].astype(int)

# converting is_canceld to category
hotel_df['is_canceled'] = hotel_df['is_canceled'].astype("category")
````

````python
# so after the data type change,the dataset would look like this
hotel_df.info()
````

**Handle Outliers in ADR (Average Daily Rate)**

````python
# Define IQR boundaries
Q1 = hotel_df['adr'].quantile(0.25)
Q3 = hotel_df['adr'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out extreme values
hotel_df = hotel_df[(hotel_df['adr'] >= lower_bound) & (hotel_df['adr'] <= upper_bound)]
````

addition of new coloumns(feature engineering)

````python
#total stay in nights
# We have created a col for total stays in nights by adding week night & weekend nights stay col.
hotel_df.loc[:, 'total_stay_in_nights'] = hotel_df['stays_in_week_nights'] + hotel_df['stays_in_weekend_nights']
````

````python
# We have created a col for revenue using total stay * ADR
hotel_df.loc[:, 'revenue'] = hotel_df['total_stay_in_nights'] * hotel_df['adr']
````

````python
#we will add a column with total number of guests coming for each booking
hotel_df.loc[:, 'total_guests'] = hotel_df['adults'] + hotel_df['children'] + hotel_df['babies']
````

removing duplicates and checking for any missing values

````python
# removing duplicates
hotel_df.drop_duplicates(inplace=True)

# To check further,if there is any null values(true/false):
hotel_df.isnull().values.any()
````

````python

# Setting the chart size
plt.figure(figsize=(15,5))

# Plotting the values in a line chart
sns.lineplot(x=bookings_by_months_df['arrival_date_month'],y=bookings_by_months_df['adr'],hue=bookings_by_months_df['hotel'])

# Setting the labels and title
plt.title('ADR across each month', fontsize=20)
plt.xlabel('Month Name', fontsize=12)
plt.ylabel('ADR', fontsize=12)

# Show chart
plt.show()
````
